```python
# Task 2

"""Scott Gleason; Thao Luu
MECH2700; Sem 2; 2009"""

from math import *
from pylab import *

"Constants Variables" 
m = 10334 #"kg"
g = 1.622 #'m/s^2'
u = 4.9 * 10 **12 #"m^3/s^2"
rm = 1.738 * 10**6 #"m"
#print "Radius of moon", rm
"Guess Value for Thrust to Mass (T/m) ratio"
Trat = 2.27226 # kgm/s^2'
print "Trust ratio is", Trat

"sets [x0,x1,x2,x3] = [r,q,dr,dq]"
"and returns dx = [dr, dq, ddr, ddq]"
def ode_set(t, z):
    r0, theta0, dr0, dtheta0 = z[0], z[1], z[2], z[3]
    
    V0 = sqrt(dr0**2 + (r0*dtheta0)**2)
    #print "Orbital Velocity =", V0, 'm/s'
    
    gama = asin(dr0/V0)
    #print "gama 0 = ", gama, 'radians'
    
    "Defines the ddr term"
    t1 = u/(r0**2)
    t2 = (Trat)*sin(gama)
    t3 = r0 * (dtheta0**2)

    ddr0 = - t1 - t2 + t3
    "Defines the ddtheta term"
    t4 = Trat
    t5 = cos(gama)/r0
    t6 = 2*dr0*dtheta0/r0

    ddtheta0 = - (t4 * t5) - t6
    dzdt = array([dr0, dtheta0, ddr0, ddtheta0])
    return dzdt

def euler_update(t, h, z0, f):
    z1 = z0 + h * f(t, z0)
    return z1

def rk4_update(t, h, z0, f):
    k1 = h * f(t, z0)
    k2 = h * f(t+h/2, z0+k1/2)
    k3 = h * f(t+h/2, z0+k2/2)
    k4 = h * f(t+h, z0+k3)
    z1 = z0 + 1.0/6*(k1 + 2*k2 + 2*k3 + k4)
    return z1
       
"Initial Values of Changing Variables"
r0 = rm + 150 * 10**3 #"m"
dr0 = 0.0 #m/s
theta0 = 0.0 #"rads"
dtheta0 = sqrt(u/(r0**3)) #rads/s
"Initail State vecotr"
z0 = array([r0, theta0, dr0, dtheta0])
t0 = 0.0
h = 0.05
dzdt=ode_set(t0, z0 )
print "dzdt=", dzdt
"Defines the [t, h, z0, f] of euler_update fuction"

x_list = []
y_list = []
for i in range(0, 100000):
#while z0[0] >= rm:
    z0 = euler_update(t0, h, z0, ode_set)
    x_list.append(z0[0]*cos(z0[1]))
    y_list.append(z0[0]*sin(z0[1]))
    if z0[0] <= rm: break

#plot the surface of the moon
x_moon = []
y_moon = []
for i in arange(0, 2*pi, pi/360):
    x_moon.append(rm*cos(i))
    y_moon.append(rm*sin(i))
plot(x_moon, y_moon, 'r')

plot(x_list, y_list)
print "Step size h is", h
print "Euler: z1=", z0
print "rk4: z1=", rk4_update(t0, h, z0, ode_set)
print "Distance landed above surface of the moon", z0[0] - rm, 'm'
fv = sqrt((z0[0]*z0[3])**2 + z0[2]**2)
print "Landing velocity is", fv, 'm/s'
#print "Landing angle is", z0[1]/pi*180, 'degrees'
show()
```
